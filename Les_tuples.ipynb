# **Les tuples**

```python
# Les tuples sont immuables (non modifiables), ordonnés et itérables.
# Autrement dit, on ne peut pas ajouter, supprimer ou modifier des éléments individuellement dans un tuple.
# Les tuples permettent de faire de la multi-assignation.
# Les tuples autorisent les doublons.
```

```python
tuple1 = (1, 2, 3, 4, 5)
print(tuple1)
```

```python
tuple1[0] = 0 # Erreur : les tuples sont immuables
```

<font color="red">NB : L'erreur ci-dessus montre que les tuples sont des objets immuables, c'est-à-dire qu'ils ne peuvent pas être modifiés une fois créés.</font>

```python
# On peut cependant réaffecter un tuple en entier
tuple1 = (0, 2, 3, 4, 5, 6) 
print(tuple1)
```

```python
# Afficher un élément d'un tuple
print(tuple1[1])
```

```python
# Créer un tuple avec des doublons
tuple2 = (1, 2, 3, 4, 5, 3, 5, 1)
print(tuple2)
```
